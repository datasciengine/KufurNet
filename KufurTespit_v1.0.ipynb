## Fonksiyon 

```python
comment = "nerden çıktın ulan acaba amk karşıma allahın belası ürün!.. orospunun çocuğu, ağzına sıçtığım."
```

```python
def get_swears(comment, black_words):
    comment = re.sub(r'[^\w\s]', '', comment)
    comment = comment.replace("  ", " ")
    comment_words = comment.split(" ")
    comment_words = [word for word in comment_words if len(word)>1]
    comment_words = [word for word in comment_words if word not in stop_words]
    
    black_list = []
    for word in comment_words:
        if word in black_words:
            black_list.append(word)
    
    len_words = len(comment_words)
    len_blacks = len(black_list)

    swear_score = round((len_blacks / len_words), 2)
    if swear_score > 0:
        return {
            "Offensive":True,
            "Score":swear_score,
            "BlackList":black_list
        }
    return {
            "Offensive":False,
            "Score":0,
            "BlackList":[]
    }
```

```python
get_swears("nerden çıktın ulan acaba amk karşıma allahın belası ürün", black_words)
```
