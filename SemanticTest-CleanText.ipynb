# E-Ticaret Yorum Veri Seti Semantik Test Deneyleri

Bu repoda temiz olduğunu düşündüğümüz e ticaret yorumlar içerisinde acaba küfür bulma ihtimalimiz var mı onu test ettik. Hipotezimiz e-ticaret yorumlarının neredeyse tamamının temiz olduğuna dayanarak çoğunda temiz bir kelimeye küfür demeyeceğimizi ispat etmek.

```python
with open("data/tr_clean_comments.txt", 'r', encoding="utf-8") as f:
    clean_comments = f.read().splitlines()
```

```python
len(clean_comments)
```

Bu deney kapsamında toplam 235.165 adet yorum analiz edilmiş oldu. Yorumların büyük çoğunluğu pozitif. Ancak biz pozitif veya negatif olma durumuyla ilgilenmiyoruz.

```python
from kufurnet import KufurNet
import pandas as pd

kufurnet = KufurNet()
```

```python
filtered_dict = {}
error_dict = {}
counter = 0

for comment in clean_comments:
    response = kufurnet.analyze({"text":comment})
    if response["status"] & response["result"]["is_black"]:
        filtered_dict[counter] = {"text":comment,
                                  "black_list":response["result"]["black_list"]}
        counter += 1

    if response["status"]==False:
        print("Something went wrong...\nComment -> ",comment, "\nResponse ->", response)
```
